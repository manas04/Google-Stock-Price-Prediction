```python
# importing necesaary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import yfinance as yf
```

# Data Retrieval of Google Stock for 5 years

```python
google = yf.Ticker('GOOG') # Taking only google data into consideration
df = google.history('5Y') #getting data for five years
df.head() # first five rows
```

# Data Visualization of the Google Stock prices

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,5))
plt.title("Closing prize overview")
plt.plot(df.Close)
```

#Addition of Moving Average to the Data-Frame

```python
df['fiftyDMA'] = df['Close'].rolling(50).mean()
df['200DMA'] = df['Close'].rolling(200).mean()
```

#Descriptive Statistics of the Data-Frame

```python
df.describe()
```

```python
df.drop(columns=['Dividends','Stock Splits'], inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,5))
plt.title("Clsoing Price vs 50 DMA vs 200 DMA")
plt.plot(df['Close'], label="Closing Price")
plt.plot(df['fiftyDMA'], label="50DMA")
plt.plot(df['200DMA'], label='200DMA')
plt.legend()
```

```python
plt.figure(figsize=(9,6))
plt.title("50 DMA")
sns.histplot(x=df['fiftyDMA'], kde = True)
```

```python
plt.figure(figsize=(9,6))
plt.title("Closing Price")
sns.histplot(x=df['Close'], kde = True)
```

```python
plt.figure(figsize=(9,6)) # Heat Map representing correlations
sns.heatmap(df.corr(), annot = True)
```

#Predicting Stock Prices Using Fitted Linear Regression Model

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols(formula = 'Close ~ fiftyDMA', data=df)
model = model.fit()
```

```python
model.summary()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,5))
plt.title("Closing Price vs fiftyDMA vs 200DMA")
plt.plot(df['Close'],label = "Closing Price")
plt.plot(df['fiftyDMA'], label = "50 DMA")
plt.plot(df['200DMA'], label = "200 DMA")
plt.legend()
```

```python
new_data = pd.DataFrame({'fiftyDMA':[105,100,103,110,115]})
predictions = model.predict(new_data)
```

```python
print(predictions)
```

#Visualization of Predicted Closing prices based on 50DMA values

```python
plt.plot(new_data, label = 'new data value')
plt.plot(predictions, label = "Predictions of the new data")
plt.legend()
```
